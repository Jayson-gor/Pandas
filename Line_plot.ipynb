```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\gorja\Downloads\salary - create a table with he data of 100 colleagues___columns_ age, dev_salary, py_salary, js_salary (1).csv")

data
```

```python
# Convert data in 'column1' from float to integer
columns_to_convert = ['Dev Salary', 'Python Salary', 'JavaScript Salary']
data[columns_to_convert] = data[columns_to_convert].astype(int)
data
```

```python
Ages = data['Age']
Developers_salary = data['Dev Salary']
Python_salary = data['Python Salary']
JavaScript_Salary = data['JavaScript Salary']

plt.plot(Ages, Developers_salary, color = 'black', linestyle = '--', label = 'Developers')
plt.plot(Ages, Python_salary, color = 'red', linestyle = '--', label = 'Python')
plt.plot(Ages, JavaScript_Salary, color = 'blue', linestyle = '--', label = 'JavaScript')

plt.legend()
plt.title('Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Monthly Salary')
plt.show()

```
