```python
import matplotlib.pyplot as plt
import numpy as np

Year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
Sales = [2000, 4000, 5000, 5500, 6500, 7000, 7800, 8700, 9300]
Profits = [500, 400, 600, 800, 1000, 1200, 1400, 1600, 1800]
Discounts = [125, 250, 375, 500, 625, 800, 910, 985, 1005]

x_indexes = np.arange(len(Year))
width = 0.3  # Adjust the width to your preference

plt.bar(x_indexes - width, Sales, width=width, color='blue', label='Sales')
plt.bar(x_indexes, Profits, width=width, color='red', label='Profits')
plt.bar(x_indexes + width, Discounts, width=width, color='black', label='Discounts')

plt.title('Sales, Profits & Discounts for Ndula')
plt.xlabel('Year')
plt.ylabel('Values')
plt.xticks(x_indexes, Year)  # Set the x-axis ticks to be the years
plt.grid(axis='y')
plt.legend()
plt.show()
```

```python
#counting t=individual langauges
from collections import Counter

language_counter = Counter()

for languages in df['LanguagesWorkedWith']:
    if pd.notna(languages):
        language_counter.update(languages.split(';'))

print(language_counter)
```

```python
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

print(languages, popularity)
```

```python
#plotting bars
plt.barh(languages, popularity)
plt.xlabel("Number of People")
plt.title("Popularity of Programming Languages in Kenya")
plt.show()
```

```python
#in descending order
languages.reverse()
popularity.reverse()
plt.barh(languages, popularity)
plt.xlabel("Number of People")
plt.title("Popularity of Programming Languages in Kenya")
plt.show()
```
