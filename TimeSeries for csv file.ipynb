```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates

# Convert 'date' column to datetime format
New_DF['date'] = pd.to_datetime(New_DF['date'])

# Set the correct year (2020) for visualization
New_DF['date'] = New_DF['date'].apply(lambda x: x.replace(year=2020))

dates = New_DF['date']
Y = New_DF['total_cases']

plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

plt.title('Covid_Deaths in 2020')
plt.xlabel('Time')
plt.ylabel('Number of Deaths')
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)

# Explicitly set x-axis ticks
plt.xticks(dates, rotation=45, ha="right")

plt.plot_date(dates, Y, linestyle='solid')
plt.show()
```
