```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")

slices = (120, 100, 80, 60, 40)
labels = ['Maths', 'English', 'Kiswahili', 'Physics', 'Business']
colors = ['blue', 'red', 'green', 'yellow', 'white']


plt.pie(slices, labels = labels, colors = colors, wedgeprops = {'edgecolor': 'black'})

plt.title("My awesome pie")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")

slices = (120, 100, 80, 60, 40)
labels = ['Maths', 'English', 'Kiswahili', 'Physics', 'Business']
colors = ['blue', 'red', 'green', 'yellow', 'white']
explode = [0,0,0,0,0.1]

plt.pie(slices, labels = labels, colors = colors,shadow = True,  startangle = 90, explode= explode, wedgeprops = {'edgecolor': 'black'})

plt.title("My awesome pie")
plt.tight_layout()
plt.show()
```
