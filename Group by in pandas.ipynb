```python
import pandas as pd

# Assuming you have already loaded your DataFrame df

# Display the unique values in the 'Station.State' column
print(df['Station.State'].unique())

# Grouping by 'Station.State'
g = df.groupby('Station.State')

# Iterating over each group
for state, state_df in g:
    print(f"State: {state}")
    print(state_df)
```

```python
# Identify numeric columns
numeric_columns = df.select_dtypes(include='number').columns

# Group by 'Station.State'
g = df.groupby('Station.State')

# Calculate mean for numeric columns
mean_df = g[numeric_columns].mean()

# Display the result
print(mean_df)
```
