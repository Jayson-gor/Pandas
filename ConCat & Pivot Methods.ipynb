```python
import pandas as pd

Kenya_weather = pd.DataFrame({
    "County": ["Nairobi", "Kisumu", "Mombasa"],
    "Temperature": [45, 22, 33],
    "Humidity": [22, 18, 15]
})

print(Kenya_weather)
```

```python
Uganda_weather = pd.DataFrame({
    "County" : ["Jinja", "Kampala", "Busia"],
    "Temperature" : [40,25,31],
    "Humidity" : [21,16,19]
})
print(Uganda_weather)
```

```python
df = pd.concat([Kenya_weather, Uganda_weather])
df
```

```python
df = pd.concat([Kenya_weather, Uganda_weather], keys=["Kenya", "Uganda"])
df
```

```python
df1 = pd.DataFrame({
    "County": ["Nairobi", "Kisumu", "Mombasa"],
    "Temperature": [45, 22, 33]
})
df1
```
