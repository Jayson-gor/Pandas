```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named oil_price_pd with columns 'TIME', 'LOCATION', and 'Value'
# and you want to plot the mean oil price for each country
mean_prices = oil_price_pd.groupby(['LOCATION', 'TIME'])['Value'].mean().unstack()

# Plotting
mean_prices.plot(kind='bar', figsize=(12, 8))
plt.title('Mean Oil Price by Country over the Years')
plt.xlabel('Country')
plt.ylabel('Mean Oil Price')
plt.legend(title='Year', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named oil_price_pd with columns 'TIME', 'LOCATION', and 'Value'
# and you want to plot the mean oil price for each country
mean_prices = oil_price_pd.groupby(['LOCATION', 'TIME'])['Value'].mean().unstack()

# Calculate the mean oil price for each country and sort them in descending order
mean_prices_mean = mean_prices.mean(axis=1).sort_values(ascending=False)
mean_prices_sorted = mean_prices.loc[mean_prices_mean.index]

# Plotting
mean_prices_sorted.plot(kind='bar', figsize=(12, 8))
plt.title('Mean Oil Price by Country over the Years (Descending Order)')
plt.xlabel('Country')
plt.ylabel('Mean Oil Price')
plt.legend(title='Year', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```
