```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")

minutes = [10, 20, 30, 40, 50, 60]
messi = [1,2,3,3,4,5]
ronaldo = [1,1,2,2,1,3]
pepe = [1,2,1,3,1,4]
labels = ['Messi', 'Ronaldo', 'Pepe']


plt.stackplot(minutes, messi, ronaldo, pepe, labels = labels)
plt.legend(loc = 'upper left')

plt.title("My awesome stackplot")
plt.tight_layout()
plt.show()
```
